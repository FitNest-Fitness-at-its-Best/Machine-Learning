```
import warnings
import os
from tqdm import tqdm
from PIL import Image

size = 512, 512

warnings.filterwarnings("ignore")

splitfolders.ratio('/content/drive/MyDrive/Dataset/dataset', output="/content/drive/MyDrive/Dataset/dataset_new/", seed=1337, ratio=(.8, .1, .1)) # default values


for folder in os.listdir('./content/drive/MyDrive/Dataset/dataset_new'):
    for directories in tqdm(os.listdir('./content/drive/MyDrive/Dataset/dataset_new/' + folder)):
        for file in os.listdir('./content/drive/MyDrive/Dataset/dataset_new/' + folder + '/' + directories):
            img = './content/drive/MyDrive/Dataset/dataset_new/' + folder + '/' + directories + '/' + file
            # print(img)
            im = Image.open(img)
            try:
                rgb_im = im.convert('RGB')
                if im.size != size:
                    # print(im.size)
                    im_resized = im.resize(size, Image.ANTIALIAS)
                    rgb_im = im_resized.convert('RGB')
                    try:
                        rgb_im.save('./content/drive/MyDrive/Dataset/dataset_new/' + folder + '/' +
                                    directories + '/' + file[:-4] + '.jpg', dpi=(72, 72))
                    except:
                        os.makedirs('./content/drive/MyDrive/Dataset/dataset_new/' + folder + '/' + directories)
                        rgb_im.save('./content/drive/MyDrive/Dataset/dataset_new/' + folder + '/' +
                                    directories + '/' + file[:-4] + '.jpg', dpi=(72, 72))
            except:
                im.close()
                os.remove(img)
```
